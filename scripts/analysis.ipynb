```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')


from utils.analysis import *
```

### Task 1
Questions and statements

```python
RESULTS_PATH = "../results"
```

```python

all_results_questions, all_results_statements = load_results(RESULTS_PATH, experiment="prompting")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.barplot(x="model", y="correct", hue="data", data=all_results_questions, ax=axes[0], legend = False).set(ylabel = "Accuracy", ylim=(0, 1))
axes[0].set_title("Questions")
axes[0].tick_params(axis="x", rotation=30)

sns.barplot(x="model", y="correct", hue="data", data=all_results_statements, ax=axes[1])
axes[1].set_title("Statements")
axes[1].tick_params(axis="x", rotation=30)
plt.suptitle("Accuracy of models on questions and statements direct prompting")
plt.tight_layout()
```
